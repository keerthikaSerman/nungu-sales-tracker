```python
import tensorflow as tf
import numpy as np

# Training data
x = np.array([
    [35, 0, 0],
    [38, 0, 1],
    [30, 1, 0],
    [40, 0, 1],
    [28, 1, 0],
    [36, 0, 0]
], dtype=float)

y = np.array([50, 90, 20, 100, 15, 60], dtype=float)

# Normalize data (IMPORTANT to avoid big wrong outputs)
x = x / 50.0
y = y / 100.0

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# Train
model.fit(x, y, epochs=500, verbose=0)

# 🔥 USER INPUT
temp = float(input("Enter temperature (°C): "))
rain = int(input("Rain? (0 = No, 1 = Yes): "))
festival = int(input("Festival? (0 = No, 1 = Yes): "))

# Normalize input
user_input = np.array([[temp, rain, festival]]) / 50.0

# Predict
prediction = model.predict(user_input)

# Convert back to real value
final_output = prediction[0][0] * 100

print("Estimated Nungu Sales:", round(final_output))
```
