```python
import math
def quantity():
  buy_quantity = int(input(f"Enter how much of nungu bought in {d}:"))
  buy_cost = int(input(f"Enter buying cost for per nungu in {d}:"))
  total_buy = buy_quantity*buy_cost
  print(f"Total cost spend for buying nungu in {d}:",total_buy)
  sale = int(input(f"Enter Total Sales of {d}:"))
  cost_day = int(input(f"Enter {d} day cost for per nungu:"))
  sale_cost = cost_day * sale
  print(f"Total sales of {d}:",sale_cost)

  if sale_cost > total_buy :
    pro = sale_cost - total_buy
    print(f"Profit {pro}")
  else:
    loss=total_buy - sale_cost
    print(f"Loss:{loss}")
  remain = buy_quantity - sale
  print(f"Remaining Stock of {d}:",remain)
print("Welcome!")
d = input("Date:")
print("1. Quantity and buying cost.")
print("2. Add total sales.")
print("3. Calculate Profit.")
print("4. Check Remaining Stack.")
print("5. Exit")
c = int(input())
if c == 1 or c==2 or c==3 or c==4:
  quantity()
elif c == 5:
  print("Exited !")
else:
  print("Invalid Chooice.")
```
